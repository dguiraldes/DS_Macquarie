Basic Forecasting using the statsmodel library - specifically using the ARIMA (AutoRegressive Integrated Moving Average) approach

```python
# Import data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from datetime import datetime, timedelta
```

```python
# Read in CPIH data frame
df_CPIH = pd.read_csv('processed_CPIH.csv')
df_CPIH.set_index('DATE', inplace=True)
```

```python
# Read in CPIH weights
df_CPIH_weights = pd.read_csv('weights_2023.csv')
df_CPIH_weights
```

```python
# Fit ARIMA models 
model1 = ARIMA(df_CPIH['01. FOOD AND NON-ALCOHOLIC BEVERAGES'], order=(24, 0, 0))
result1 = model1.fit()

model2 = ARIMA(df_CPIH['02.ALCOHOLIC BEVERAGES,TOBACCO & NARCOTICS'], order=(24, 0, 0))
result2 = model2.fit()

model3 = ARIMA(df_CPIH['03. CLOTHING AND FOOTWEAR'], order=(24, 0, 0))
result3 = model3.fit()

model4 = ARIMA(df_CPIH['04. HOUSING, WATER, ELECTRICITY, GAS AND OTHER FUELS'], order=(24, 0, 0))
result4 = model4.fit()

model5 = ARIMA(df_CPIH['05. FURN, HH EQUIP & ROUTINE REPAIR OF HOUSE'], order=(24, 0, 0))
result5 = model5.fit()

model6 = ARIMA(df_CPIH['06. HEALTH'], order=(24, 0, 0))
result6 = model6.fit()

model7 = ARIMA(df_CPIH['07. TRANSPORT'], order=(24, 0, 0))
result7 = model7.fit()

model8 = ARIMA(df_CPIH['08. COMMUNICATION'], order=(24, 0, 0))
result8 = model8.fit()

model9 = ARIMA(df_CPIH['09. RECREATION & CULTURE'], order=(24, 0, 0))
result9 = model9.fit()

model10 = ARIMA(df_CPIH['10. EDUCATION'], order=(24, 0, 0))
result10 = model10.fit()

model11 = ARIMA(df_CPIH['11. HOTELS, CAFES AND RESTAURANTS'], order=(24, 0, 0))
result11 = model11.fit()

model12 = ARIMA(df_CPIH['12. MISCELLANEOUS GOODS AND SERVICES'], order=(24, 0, 0))
result12 = model12.fit()
```

```python
# Forecast
forecast_steps = 3  # Forecasting 3 months ahead
forecast1 = result1.forecast(steps=forecast_steps)  
forecast2 = result2.forecast(steps=forecast_steps)  
forecast3 = result3.forecast(steps=forecast_steps)  
forecast4 = result4.forecast(steps=forecast_steps)  
forecast5 = result5.forecast(steps=forecast_steps)  
forecast6 = result6.forecast(steps=forecast_steps)  
forecast7 = result7.forecast(steps=forecast_steps)  
forecast8 = result8.forecast(steps=forecast_steps)  
forecast9 = result9.forecast(steps=forecast_steps)  
forecast10 = result10.forecast(steps=forecast_steps)  
forecast11 = result11.forecast(steps=forecast_steps)  
forecast12 = result12.forecast(steps=forecast_steps)  
```
